```python
from pylab import *
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data-AiteP.csv')
```

```python
x = df.year
y1 = df.HKP
y2 = df.SingaporeP
y3 = df.HKR
y4 = df.SingaporeR
y = [y1,y2,y3,y4]
```

```python
fig, ax = plt.subplots()

plt.ylim(0,24.8)

ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')

plt.plot(x,y1,'bo--',label='HKP')
plt.plot(x,y2,'go--',label='SingaporeP')
plt.plot(x,y3,'bo-',label='HKR')
plt.plot(x,y4,'go-')

tick_params(top='off',bottom='off',left='off',right='off')


plt.grid(axis="y")

plt.legend(bbox_to_anchor=(0, 1, 1, 0),mode="expand", borderaxespad=0)
plt.show()
```

```python
df = pd.read_csv("data-YBswW.csv")
```

```python
x1 = df.Way
y1 = df.Number
```

```python
from matplotlib.patches import Wedge
import matplotlib.pyplot as plt
size = 1
#创建一个画板
fig,ax=plt.subplots()

#创建半圆

plt.pie(y1,radius=2,
       wedgeprops = dict(width = size,edgecolor='w'),
       labels = x1)
semicircle=Wedge((0,0),2,200,340)
semicircle.set_zorder(1)
ax.add_patch(semicircle)
plt.axis('equal')
```

```python
df = pd.read_csv("data-DWh55.csv")
```

```python
x = df.Year
y = df.Number
```

```python
fig, ax = plt.subplots()

ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')

x_major_locator = MultipleLocator(1)
y_major_locator = MultipleLocator(50)
ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(y_major_locator)

ax.set_title('Plastic dishware waste at landfills in HK')

left = -0.15
bottom, height = .25, .75
top = bottom + height

ax.text(left, top, 'Unit: tonnes per day',
        horizontalalignment='left',
        verticalalignment='top',
        transform=ax.transAxes)

plt.grid(axis="y")
plt.rc('axes', axisbelow=True)

tick_params(top='off',bottom='off',left='off',right='off')

plt.bar(x,y,width=0.7)

```
